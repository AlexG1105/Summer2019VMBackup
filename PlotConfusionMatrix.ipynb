```python
def normalize_row_and_columns(cm):
    for i in range(0,cm[0].size - 1):
        total = np.sum(array[i:,i]) + np.sum(array[i,i+1:])
        for j in range(i, cm[0].size):
            cm[j,i] = cm[j,i] / total
        for k in range(i+1, cm[0].size):
            cm[i,k] = cm[i,k] / total
    return cm
```

```python
normalarray = normalize_row_and_columns(array)
```

```python
#Normalized random undersampled object detection model
df_cm = pd.DataFrame(normalarray, index = ["Accent Chairs","Area Rugs", "Coffee Tables", "Console Tables", "End Tables", "Floor Lamps", "Ottomans", "Sofas and Loveseats", "Table Lamps", "Throw Pillows", "Vases", "No Prediction"],
                  columns = ["Accent Chairs","Area Rugs", "Coffee Tables", "Console Tables", "End Tables", "Floor Lamps", "Ottomans", "Sofas and Loveseats", "Table Lamps", "Throw Pillows", "Vases", "No Prediction"])

plt.figure(figsize = (13,10))
sn.set(font_scale=1.5)
ax = sn.heatmap(df_cm, annot=True, cmap="Blues",annot_kws={"size": 14})
ax.set(xlabel='Predicted', ylabel='Actual')
```
