```python
import pandas as pd
import os
import shutil
```

```python
#start off with just product images
attributeDF = pd.read_csv('Osamadata/TableLampsBaseMaterialData/finaltablelamps_basematerial.csv')
images_path = 'Osamadata/TableLampsBaseMaterialData/TableLampsBaseMaterialProductImages'
image_extensions = ['.jpg', '.png', '.jpeg']
images = [os.path.join(dp, f) for dp, dn, filenames in os.walk(images_path)
         for f in filenames if os.path.splitext(f)[1].lower() in image_extensions]
print(len(images))
```

```python
attributeDF.groupby(['attributevalue'])['displayname'].count()
```

```python
attributeDF['cc'] = attributeDF.groupby(['attributevalue']).cumcount()
attributeDF = attributeDF.loc[attributeDF['cc'] < 125]
```

```python
data = 'Osamadata/basematerial'
keeplist = ['Ceramic','Crystal','Glass','Iron','Metal','Polyresin','Resin','Steel']
for i,image in enumerate(images):
    ID = image.split('/')[-1].split('.')[0]
    row = attributeDF.loc[attributeDF['omsid'] == int(ID)]
    
    if (row.shape[0] == 0):
        continue
    
    print(row.shape[0])
    
    for item in row.itertuples():
        attribute = item.attributevalue
        
    if attribute not in keeplist:
        continue
    
    if i % 10 == 1:
        split = 'valid'
    else:
        split = 'train'
    
    path = os.path.join(data,split,attribute)
    if not os.path.isdir(path):
        os.makedirs(path)
    
    shutil.copy(image, path + '/' + image.split('/')[-1])
    
```

```python
##now we add in cropped
croppedImages = []
images_path = 'Osamadata/TableLampsBaseMaterialData/TableLampsCropped'
for dp, dn, filenames in os.walk(images_path):
    for f in filenames:
        croppedImages.append(os.path.join(dp,f))
croppedIDs = [image.split('/')[-1].split('.')[0] for image in croppedImages]
croppedIDs = [int(item) for item in croppedIDs if item != '']
print(len(croppedIDs))
```

```python
#cropped count
attributeDF.loc[attributeDF['omsid'].isin(croppedIDs)].groupby(['attributevalue'])['displayname'].count()
```

```python
valattDF = attributeDF.loc[attributeDF['omsid'].isin(croppedIDs)]
valattDF['cc2'] = valattDF.groupby(['attributevalue']).cumcount()
valattDF = valattDF.loc[valattDF['cc2'] < 50]
```

```python
allImages = images + croppedImages
```
