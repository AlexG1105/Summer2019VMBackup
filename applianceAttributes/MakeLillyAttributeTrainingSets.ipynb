```python
#Now we want to start creating symbolic links in this structure:
#category/attribute/train|valid/label/symlinks
path = 'lrdata/sofa/material/'
searchpath = 'LivingRoomImages/'
for i,item in enumerate(d14.drop_duplicates(['product_id']).itertuples()):
    if (i % 10 == 0):
        phase = 'valid'
    else:
        phase = 'train'

    if not os.path.exists(path + phase):
        os.makedirs(path + phase)
        
    category = item.material
    if '/' in category:
        category = category.replace('/','.')
        print(category)
    
    if not os.path.exists(path + phase + '/' + category):
        os.makedirs(path + phase + '/' + category)
        
    hashnum = item.product_id % 100
    
    os.symlink(searchpath + str(hashnum) + '/' + str(item.product_id) + '.jpg',
               path + phase + '/' + category + '/' + str(item.product_id) + '.jpg')
```
